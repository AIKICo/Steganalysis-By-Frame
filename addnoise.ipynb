```python
import numpy as np
import random
import itertools
import librosa
import IPython.display as ipd
import matplotlib.pyplot as plt
import soundfile as sf

%matplotlib inline
```

```python
def load_audio_file(file_path):
    data, sr = librosa.core.load(file_path)
    print(sr)
    return data, sr

def plot_time_series(data):
    fig = plt.figure(figsize=(14, 8))
    plt.title('Raw wave ')
    plt.ylabel('Amplitude')
    plt.plot(np.linspace(0, 1, len(data)), data)
    plt.show()
```

```python
data, sr = load_audio_file("SteganalysisDatasets/Normal/PDAm01_009_5.wav")
plot_time_series(data)
```

```python
ipd.Audio(data, rate=22050)
```

```python
wn = np.random.randn(len(data))
data_wn = data + 0.001*wn
plot_time_series(data_wn)
# We limited the amplitude of the noise so we can still hear the word even with the noise, 
#which is the objective
ipd.Audio(data_wn, rate=22050)
```

```python
 # librosa.output.write_wav('SteganalysisDatasets/NoiseData/NormalNoise/PDAm01_009_4.wav', data_wn, sr)
sf.write('SteganalysisDatasets/NoiseData/NormalNoise/PDAm01_009_5.wav', data_wn, 22050, 'PCM_16')
```
